```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("UBC-NLP/AraT5-base-title-generation")

model = AutoModelForSeq2SeqLM.from_pretrained("UBC-NLP/AraT5-base-title-generation")
```

```python
import pandas as pd

data = pd.read_csv("data.csv")
```

```python
from tqdm import tqdm
```

```python
new_titles = []
for text in tqdm(data["raw_texts"]):
    input_ids = tokenizer.encode(text, return_tensors="pt")
    outputs = model.generate(input_ids)
    new_titles.append(tokenizer.decode(outputs[0], skip_special_tokens=True,clean_up_tokenization_spaces=True))
```

```python
new_titles
```

```python
data["titles"] = new_titles

data.to_csv("data.csv", index=False)
```
